```python
praha = pandas.read_csv("zam_praha.csv")
plzen = pandas.read_csv("zam_plzeň.csv")
liberec = pandas.read_csv("zam_liberec.csv")

# pro moji informaci:
celkem = len(praha) + len(plzen) + len (liberec)
celkem
```

```python
# Ke každé tabulce přidej nový sloupec mesto, které bude obsahovat informaci o tom, ve kterém městě zaměstnanec pracuje.
praha['lokace'] = 'praha'
plzen['lokace'] = 'plzen'
liberec['lokace'] = 'liberec'

# Vytvoř novou tabulku zamestnanci a ulož do ní informace o všech zaměstnancích.
zamestnanci = pandas.concat([praha, plzen, liberec], ignore_index=True)
zamestnanci.to_csv('zamestnanci.csv', index=False)
# kontrolně:
if len(zamestnanci) == zamestnanci["cislo_zamestnance"].nunique():
    print("cisla_zamestnance jse neopakují")
else:
    print("některá cisla_zamestnance jsou víckrát")
zamestnanci. head(10)
```

```python
# načti platy zaměstnanců za únor
platy_21_02 = pandas.read_csv("platy_2021_02.csv")
platy_21_02.head()
```

```python
# Propoj tabulku (operace join) s platy a tabulku se zaměstnanci pomocí sloupce cislo_zamestnance.

propojeny_df = pandas.merge(zamestnanci, platy_21_02, how="left" )

# porovnej rozměry tabulek před spojením a po spojení
if len(propojeny_df) == len(zamestnanci):
    print('Počet řádků stejný')
else:
    print('změnil se počet řádků!!')
propojeny_df
```

```python
# Pokud nemá některý zaměstnanec plat za únor, znamená to, že v naší firmě již nepracuje.
zivi =  propojeny_df[propojeny_df['plat'].notnull()]
zivych = zivi["cislo_zamestnance"].nunique()
print(f"Živých zaměstnanců máme {zivych}.")
zivi.head()
```

```python
# Spočti průměrný plat zaměstnanců v jednotlivých kancelářích.
zivi.groupby('lokace')['plat'].mean()

# a kdybych chtěla průměr za všechny, tak stačí
# zivi['plat'].mean()
```

```python
# Ulož do proměnné počet zaměstnaců, kteří v naší firmě již nepracuji
poc_neprac = len(propojeny_df[propojeny_df['plat'].isnull()])
poc_neprac
```

```python
# Vytvoř samostatnou tabulku, která obsahuje jména zaměstnanců, kteří ve firmě již nepracují. Tabulku ulož do souboru CSV.
nepracuji = propojeny_df[propojeny_df['plat'].isnull()]
# nepracuji

nepracuji.to_csv('nepracuji.csv', index=False)
nepracuji
```

```python
import pytemperature
mereni = pandas.read_csv("temperature.csv")
mereni["AvgTemperatureCelsia"] = pytemperature.f2c(mereni["AvgTemperature"])
mereni
```

```python
# Dotaz na řádky z 13. listopadu 2017 ze Spojených států amerických
mereni13US = mereni.query("Country == 'US' and Day == 13")
mereni13US
```

```python
#  data z předchozího dotazu napiš dotaz na řádky ve městech (sloupec City) Washington a Philadelphia.
mereni13towns = mereni13US.query("City == 'Philadelphia' or City == 'Washington'")
mereni13towns

# to už bylo ale všechno minule, ne?
```

```python
# zpět k pomocné tabulce, kterou jsi vytvořila v bodu 2. Vypiš průměrnou hodnotu ze všech měření, 
# která byla provedena 13. listopadu 2017 na území Spojených států amerických. K tomu využij funkci .mean(), 
# která funguje stejně jako funkce .sum(), kterou jsme si ukazovali na lekci. 

prumer13US = mereni13US['AvgTemperatureCelsia'].mean()
prumer13US
# Pokud znáš základy statistiky, zkus funkci pro medián .median() a rozptyl .var().
```

```python
median13US = mereni13US['AvgTemperatureCelsia'].median()
median13US
```

```python
rozptyl13US = mereni13US['AvgTemperatureCelsia'].var()
rozptyl13US
```
