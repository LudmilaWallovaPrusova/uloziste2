```python
import pytemperature
mereni["AvgTemperatureCelsia"] = pytemperature.f2c(mereni["AvgTemperature"])
mereni.head()
```

```python
# Dotaz na měření, ve kterých je teplota (sloupec AvgTemperatureCelsia) vyšší než 30 stupňů Celsia.
mereni.query("AvgTemperatureCelsia > 30").head()
```

```python
# Dotaz na měření, ve kterých je teplota vyšší než 17 stupňů Celsia a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).
mereni.query("AvgTemperatureCelsia > 17 and Region == 'Europe'")
```

```python
# Dotaz na měření, ve kterých je teplota vyšší než 15 stupňů Celsia a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).
EvropaNad17C = mereni[(mereni["AvgTemperatureCelsia"] > 17) | (mereni["Region"] == "Europe")]
EvropaNad17C
```

```python
# Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 30 stupňů Celsia nebo menší než -10 stupňů.
mereni.query("AvgTemperatureCelsia > 30 or AvgTemperatureCelsia < -10")
```
